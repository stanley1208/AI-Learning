```python
import numpy as np
import matplotlib.pyplot as plt



x=np.linspace(0,10,50)
y=2*x+1+np.random.randn(50)
```

```python
m,b=0.0,0.0
lr=0.00001
epochs=100
```

```python
losses=[]
```

```python
for i in range(epochs):
  y_pred=m*x+b
  error=y-y_pred
  loss=np.mean(error**2)
  losses.append(loss)

  dm=2*np.mean(error*x)
  db=2*np.mean(loss)

  m-=lr*dm
  b-=lr*db

```

```python
print(f"learned parameters: m={m:.2f}, b={b:.2f}")
print("final loss:",round(losses[-1],3))
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.scatter(x,y,label="Data")
plt.plot(x,m*x+b,color='red',label='learned line')
plt.legend()
plt.title("model fit")

plt.subplot(1,2,2)
plt.plot(losses)
plt.title("losses over epochs")
plt.xlabel('epoch')
plt.ylabel('MSE')
plt.tight_layout()
plt.show()
```
