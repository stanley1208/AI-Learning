```python
import torch
import math
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torch.nn.functional as F
from torchvision import datasets, transforms
from torchvision.models import resnet50,ResNet50_Weights
from torchvision.models import resnet18,ResNet18_Weights
import matplotlib.pyplot as plt
```

```python
def scaled_dot_product_attention(Q,K,V):
  """
    Q: [B, L, d]
    K: [B, L, d]
    V: [B, L, d]

    Returns:
        output: [B, L, d]
        attn_weights: [B, L, L]
    """
  scores=torch.matmul(Q,K.transpose(-2,-1))   # [B,L,L]
  scores=scores/math.sqrt(Q.size(-1))         # scale by âˆšd
  attn_weights=torch.softmax(scores,dim=-1)   # softmax over keys
  output=torch.matmul(attn_weights,V)         # [B, L, d]
  return output,attn_weights
```

```python
B,L,D=2,4,8     # small numbers on purpose
Q=torch.randn(B,L,D)
K=torch.randn(B,L,D)
V=torch.randn(B,L,D)

```

```python
out, attn = scaled_dot_product_attention(Q, K, V)
print("Q:", Q.shape, "K:", K.shape, "V:", V.shape)
print("out:", out.shape)
print("attn:", attn.shape)
```
