```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,50)
y=2*x+1+np.random.randn(50)

X=np.c_[x,np.ones(x.shape[0])]
```

```python
theta=np.zeros(2)
```

```python
def compute_loss(X,y,theta):
  error=X@theta-y
  return np.mean(error**2)

def compute_grad(X,y,theta):
  error=X@theta-y
  return (2/(len(y)))*(X.T@error)
```

```python
lrs=[0.00001,0.0001,0.01]
epochs=200

plt.figure(figsize=(10,4))
for lr in lrs:
  theta=np.zeros(2)
  losses=[]

  for i in range(epochs):
    grad=compute_grad(X,y,theta)
    theta-=lr*grad
    losses.append(compute_loss(X,y,theta))

  plt.plot(losses,label=f"lr={lr}")
  print(f"LR={lr} -> final theta={theta.round(3)}, Final Loss={round(losses[-1],4)}")



plt.title("loss curves for different learning rates")
plt.xlabel("epochs")
plt.ylabel("MSE")
plt.legend()
plt.show()
```
